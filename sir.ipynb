```python
# Encode categorical variables
tips_encoded = pd.get_dummies(tips, drop_first=True)
# Calculate the correlation matrix
tips_corr = tips_encoded.corr()
# Plot the heatmap
sns.heatmap(tips_corr, cmap='coolwarm', annot=True)
```
